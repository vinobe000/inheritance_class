```python
class vinoth():
  def __init__(self,name,age,city):
    self.name=name
    self.age=age
    self.city=city
    
```

```python
class kumar(vinoth):
  def display(self):
    print(self.name)
    print(self.age)
    print(self.city)
    
```

```python
kamblee=kumar("vinoth",28,"salem")
```

```python
kamblee.display()
```
